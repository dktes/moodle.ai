# Scikit-learn implementation
sckit-learn is a very powerful library for data-science. The complete code is given below

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
# generate random data-set
np.random.seed(0)
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.rand(100, 1)
```

```python
# model initialization
lr = LinearRegression()

# model fit
lr.fit(x,y)

# Predict
y_predicted = lr.predict(x)

# model evaluation
rmse = math.sqrt(mean_squared_error(y, y_predicted))
```

```python
print('Slope:' ,lr.coef_)
print('Intercept:', lr.intercept_)
print('Root mean squared error: ', rmse)
```

This is almost similar to what we achieved when we implemented linear regression from scratch.

```python
# plotting values

# data points
plt.scatter(x, y, s=10)
plt.xlabel('x')
plt.ylabel('y')

# predicted values
plt.plot(x, y_predicted, color='r')
```
